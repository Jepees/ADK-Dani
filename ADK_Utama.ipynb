## Relative Risk (RR)

```python
beda_proporsi = pi_1 - pi_2

print("Beda Proporsi :")
print(f" = {beda_proporsi}")
```

## Likelihood Ratio Test

```python
def stat_G2(kontigen, fr_harapan):
    a = kontingen.to_numpy()
    b = fr_harapan.to_numpy()

    G2 = 0
    for i in range(len(a)):
        for j in range(len(a[i])):
            G2 += a[i][j] * np.log(a[i][j] / b[i][j])

    return 2*G2
```

## CMH Test

```python
def stat_CMH(Z1, Z2):
    E1 = (Z1.loc[Z1.index[0]].sum()) * Z1[Z1.columns[0]].sum() / Z1.sum().sum()
    E2 = (Z2.loc[Z2.index[0]].sum()) * Z2[Z2.columns[0]].sum() / Z2.sum().sum()

    var1 = Z1.transpose().sum()[0] * Z1.transpose().sum()[1] * Z1.sum()[0] * Z1.sum()[1] / (((Z1.sum().sum())**2) * (Z1.sum().sum() - 1))
    var2 = Z2.transpose().sum()[0] * Z2.transpose().sum()[1] * Z2.sum()[0] * Z2.sum()[1] / (((Z2.sum().sum())**2) * (Z2.sum().sum() - 1))

    hasil = ((Z1.loc[Z1.index[0], Z1.columns[0]] - E1) + (Z2.loc[Z2.index[0], Z2.columns[0]] - E2)) / np.sqrt(var1 + var2)
    return hasil

stat_CMH  = stat_CMH(Parsial_Z1, Parsial_Z2)
```

```python
print("statistik uji CMH (X^2)")
print(f" = {stat_CMH}")
```

```python
print("p value CMH test")
print(f"p-value = 2*P(Z > stat_CMH)")
print(f"p-value = 2*P(Z > {stat_CMH})")
print(f"p-value = 2*({norm.sf(stat_CMH)}))")
print(f'p-value = {2 * norm.sf(stat_CMH)}')
```

```python
def est_common_OR(Z1, Z2):
    a1 = Z1.loc[Z1.index[0], Z1.columns[0]] * Z1.loc[Z1.index[1], Z1.columns[1]] / Z1.sum().sum()
    a2 = Z2.loc[Z2.index[0], Z2.columns[0]] * Z2.loc[Z2.index[1], Z2.columns[1]] / Z2.sum().sum()

    b1 = Z1.loc[Z1.index[0], Z1.columns[1]] * Z1.loc[Z1.index[1], Z1.columns[0]] / Z1.sum().sum()
    b2 = Z2.loc[Z2.index[0], Z2.columns[1]] * Z2.loc[Z2.index[1], Z2.columns[0]] / Z2.sum().sum()

    hasil = (a1 + a2) / (b1 + b2)

    return hasil
```

## Evaluasi Perbandingan Antara 2 Model

```python
lr_statistic = model1.deviance - model2.deviance # selisih 2 deviance
degree_of_freedom = X1_train.shape[1] - X2_train.shape[1]  # Derajat kebebasan (jumlah parameter tambahan)

# Menghitung p-value untuk uji LR
p_value = chi2.sf(lr_statistic, abs(degree_of_freedom))
print(f"LR Statistic: {lr_statistic}")
print(f"P-Value: {p_value}")
```

```python
print(f"dengan LR Statistic: {lr_statistic}")
print(f"dan P-Value: {p_value}")
if p_value < 0.05:
    print(f"karena p_value < 0.05\nmaka model yang lebih kompleks lebih baik\nartinya model dengan interaksi antara Age dengan {X_catt} lebih baik")
else:
    print(f"karena p_value > 0.05\nmaka model yang lebih sederhana lebih baik\nartinya model tanpa interaksi antara Age dengan {X_catt} lebih baik")
```
